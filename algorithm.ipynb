```python
import pandas as pd
from glob import glob
```

```python
close_table = pd.DataFrame()
for file_name in glob(r'./data/*.csv'):
    df = pd.read_csv(file_name)
    ticker = file_name.split('/')[-1].split('.')[0]
    close_table[ticker] = df['adjclose']
close_table.head()
```

```python
x = close_table.pct_change().fillna(0).iloc[1:, :].transpose()
```

```python
from sklearn.cluster import KMeans
```

```python
k = KMeans(n_clusters=10, random_state=0).fit_predict(x)
```
